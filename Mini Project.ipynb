```python
#lets use a function
def missing_num(x):
    return sum(x.isnull())

print('Missing values per column')
#applying our function by column
#check sum of missing values by column, and get the columns where 
#the sum is not zero

print(online_retail.apply(missing_num, axis = 0))
```

```python
#lamda arguments:expression
#it is used to print the missing values only. It doesnt print the column without a missing value
print('The missing value without having zero')
print(online_retail.apply(missing_num, axis = 0).where(lambda x: x!=0).dropna())
```

```python
#To get total sales trend per month we need to:
#Convert the date column to date time object for it to work.
time_col = 'InvoiceDate'
online_retail[time_col] = pd.to_datetime(online_retail[time_col])
online_retail.InvoiceDate
```

```python
time_data = online_retail.resample ('M', on = time_col)['Totalsale'].sum().fillna(0)
time_data
fig, ax = plt.subplots(figsize = (16,6))
ax.plot(time_data)
plt.title('Total Sales Trend from 2010-2012 per month')
```
